```python
import numpy as np
import os
import matplotlib.pyplot as plt

# EXTRACT HOME FOLDER
HOME = '/home/jacopo'

# INITIALIZE EXPLAINERS DICTIONARY
expmapper = {
  0: {'name': 'H-Explainer'},
  1: {'name': 'GradCAM'},
  2: {'name': 'GradCAM++'}
}
EXP_L = len(expmapper) # number of explainers

# IMPORT PREVIOUSLY COMPUTED LOR MAP
LOR = np.load(os.path.join(HOME, 'repo/hshap/data/rsna/LOR/results/no_shap_300_1600_10.npy'))
exp_x = np.linspace(-1, 0, 20)
perturbation_sizes = np.sort(1.1 - 10**(exp_x))
perturbations_L = len(perturbation_sizes)
LOR_shape = LOR.shape
print(LOR.shape)

EXP_L = LOR_shape[0]
IMG_L = LOR_shape[1]
PERTURBATION_L = LOR_shape[2]
```

```python
# DEFINE OUTPUT IMAGE GRID
ref_sizes = [100, 200, 400, 800, 1600]
fig = plt.figure(figsize=(10, 5 * len(ref_sizes) + 2))
axes = fig.subplots(len(ref_sizes), 1)

for i, ref_size in enumerate(ref_sizes):
    
    LOR = np.load(os.path.join(HOME, 'repo/hshap/data/rsna/LOR/results/no_shap_300_{}_10.npy'.format(ref_size)))
    
    ax = axes[i]
    ax.set_xlabel('Perturbation size')
    ax.set_ylabel('LOR')
    ax.set_title("Reference size = {}".format(ref_size))

    for exp_id in np.arange(EXP_L):
        exp_name = expmapper[exp_id]['name']
        ax.plot(perturbation_sizes, np.mean(LOR[exp_id, :, :], axis = 0), label = exp_name)
        ax.legend()
```
