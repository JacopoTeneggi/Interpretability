```python
import numpy as np
import os
import matplotlib.pyplot as plt

# EXTRACT HOME FOLDER
HOME = '/home-net/home-2/jtenegg1@jhu.edu'

# INITIALIZE EXPLAINERS DICTIONARY
expmapper = {
  0: {'name': 'Grad-Explainer'},
  1: {'name': 'Deep-Explainer'},
  2: {'name': 'H-Explainer'},
  3: {'name': 'GradCAM'},
  4: {'name': 'GradCAM++'},
  5: {'name': 'Kernel-SHAP'}
}
EXP_L = len(expmapper) # number of explainers

# IMPORT PREVIOUSLY COMPUTED LOR MAP
LOR = np.load(os.path.join(HOME, 'data/Jacopo/HShap/LOR/perturbation.npy'))
perturbation_sizes = np.linspace(0, 1, 100)
LOR_shape = LOR.shape
print(LOR.shape)

CROSSES_L = LOR_shape[0]
EXP_L = LOR_shape[1]
IMG_L = LOR_shape[2]
PERTURBATION_L = LOR_shape[3]
```

```python
# DEFINE OUTPUT IMAGE GRID
fig = plt.figure(figsize=(20, 4))
axes = fig.subplots(1, CROSSES_L)

for crosses in np.arange(CROSSES_L):
  
  ax = axes[crosses]
  ax.set_xlabel('Perturbation size')
  ax.set_ylabel('LOR')
  ax.set_title('Images with %d crosses' % (crosses + 1))

  for exp_id in np.arange(EXP_L):

    exp_name = expmapper[exp_id]['name']
    ax.plot(perturbation_sizes, np.mean(LOR[crosses, exp_id, :, :], axis = 0), label = exp_name)
    ax.legend()
```
