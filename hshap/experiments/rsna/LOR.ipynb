```python
import numpy as np
import os
import matplotlib.pyplot as plt

# EXTRACT HOME FOLDER
HOME = '/home/jacopo'

# INITIALIZE EXPLAINERS DICTIONARY
expmapper = {
  0: {'name': 'Grad-Explainer'},
  1: {'name': 'Deep-Explainer'},
  2: {'name': 'H-Explainer'},
  3: {'name': 'GradCAM'},
  4: {'name': 'GradCAM++'}
}
EXP_L = len(expmapper) # number of explainers

# IMPORT PREVIOUSLY COMPUTED LOR MAP
LOR = np.load(os.path.join(HOME, 'repo/hshap/data/rsna/LOR_same_baseline.npy'))
perturbation_sizes = np.linspace(.10, 1, 10)
LOR_shape = LOR.shape
print(LOR.shape)

EXP_L = LOR_shape[0]
IMG_L = LOR_shape[1]
PERTURBATION_L = LOR_shape[2]
```

```python
# DEFINE OUTPUT IMAGE GRID
fig = plt.figure(figsize=(20, 4))
ax = fig.subplots(1)
ax.set_xlabel('Perturbation size')
ax.set_ylabel('LOR')

for exp_id in np.arange(EXP_L):
    exp_name = expmapper[exp_id]['name']
    ax.plot(perturbation_sizes, np.mean(LOR[exp_id, :, :], axis = 0), label = exp_name)
    ax.legend()
```
