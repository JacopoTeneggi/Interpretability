```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import numpy as np
import pandas as pd
from tqdm import tqdm

c = [1, 6]
exp_mapper = ["hexp/absolute_0", "gradexp", "deepexp", "partexp", "lime"]
LOR_df_clips = []
for j, n in enumerate(c):
    for exp_name in tqdm(exp_mapper):
        exp_df = pd.read_csv(os.path.join("LOR", exp_name, f"results_{n}.csv"))
        LOR_df_clips.append(exp_df)
LOR_df = pd.concat(LOR_df_clips, axis=0)
LOR_df["perturbation_size"] /= 100*120
```

```python
sns.set_theme()
sns.set_context("paper")

g = sns.FacetGrid(LOR_df, col="n", hue="exp_name", height=4)
g.map(sns.lineplot, "perturbation_size", "logit")
g.set(xlabel="Fraction of image ablated", ylabel="Logit output as a function of original logit")
g.add_legend(title="Explainer")
plt.savefig("LOR_seaborn.eps")
plt.savefig("LOR_seaborn.jpg")
```
